```python
#1

#Using the indicator calculated previously: Total Population
#Filter the 'data' to include only countries with a total population between 9000000 and 11000000
populated_countries = data[data["POPULATION"].between(9000000, 11000000)]

populated_countries
```

```python
#2

#Using the indicator calculated previously: GDP measured in US Dolars
#Calculate the average of GDP for the countries in 'populated_countries' and store it in 'average_gdp'
average_gdp = populated_countries["GDP"].mean()

#Calculate the standard deviation of GDP for the countries in 'populated_countries' and store it in 'std_dev_gdp'
std_dev_gdp = populated_countries["GDP"].std()

print(f"The average of gdp for this countries is {average_gdp}")
print(f"The standard deviation in the GDP is {std_dev_gdp}")
```

```python
#3

#Assuming we want the indicator GDP per capita currents US dolars
#Calculate the average GDP per capita for the countries in 'populated_countries' and store it in 'average_gdppercapita'
average_gdppercapita = np.mean((populated_countries['GDP'])/populated_countries['POPULATION'])

#Calculate the standard deviation of GDP per capita for the countries in 'populated_countries' and store it in 'std_dev_gdppercapita'
std_dev_gdppercapita = np.std((populated_countries['GDP'])/populated_countries['POPULATION'])

print(f"The average of gdp per capita for this countries is {average_gdppercapita}")
print(f"The standard deviation in the GDP per capita is {std_dev_gdppercapita}")
```

```python
#4

#Create the dictionary the will classifify the size of the country
classifier = {'Tiny': 0, 'Very Small': 1000000, 'Small': 5000000, 'Medium': 15000000,'Large': 30000000, 'Huge': 100000000}

#Create a new column with the values of the GDP per capita for each country
data['GDP per Capita'] = data['GDP'] / data['POPULATION']

#Introducing a new column that will store the classification of each country size, using the dictionary created
#"np.inf" is used to handle any population values that are larger than the highest range of the bins
data['Size Classification'] = pd.cut(data['POPULATION'], bins=[classifier[size] for size in classifier.keys()] + [np.inf], labels=classifier.keys())

#Create groups with the size classifications and calculate the mean and std of the GDP per capita of each one
data.groupby('Size Classification')['GDP per Capita'].agg(['mean', 'std'])
```

```python
#5
#Apply a lambda function that returns 1 if value is above the mean, 0 otherwise, and NaN if the value is missing
def above_mean(df, col):
    return df[col].apply(lambda x: 1 if not np.isnan(x) and x > df[col].mean() else (0 if not np.isnan(x) else np.nan))
print(above_mean(data,'GINI'))
```

```python
#6
#Create a new column 'hc_above_average' to indicate if human capital ('hc') is above the average, returning 0 or 1 accordingly
data['hc_above_average'] = above_mean(data,'hc')

#Group by 'Size Classification' and 'hc_above_average', then calculate the average GDP per Capita for each group
data.groupby(['Size Classification', 'hc_above_average'])[['GDP per Capita']].mean().reset_index()
```

```python
#7 
#Create a new column 'gini_above_average' to indicate if the Gini is above the average, returning 0 or 1 accordingly
data['gini_above_average']=above_mean(data,'GINI')

#Group by whether human capital and GINI are above or below average, and calculate the average GDP per capita for each group
#Reset the index to make it more organized
data.groupby(['hc_above_average','gini_above_average'])[['GDP per Capita']].mean().reset_index()
```

```python
#8

#Filter the data for countries with both human capital and Gini coefficient below average (0)
data_filtered = (data['hc_above_average'] == 0) & (data['gini_above_average'] == 0)

#Find the index (country code) of the country with the highest GDP per capita among the filtered data
country_code = data.loc[data_filtered, 'GDP per Capita'].idxmax()

#Printing the name of that country
print(wdi[wdi['Country Code'] == country_code]['Country Name'].iloc[0])
```

```python
#9 

#Create a series with the mean of the GINI for each size classification
average_gini = data.groupby('Size Classification')['GINI'].transform('mean')

#Create a series with the mean of the human capital (hc) for each size classification
average_human_capital = data.groupby('Size Classification')['hc'].transform('mean')

#Filter the data for countries where both criteria is satisfied
countries_filter = data[(data['GINI'] < average_gini) & (data['hc'] < average_human_capital)]

#Find the row with the highest GDP per capita from the filtered data
max_country= countries_filter.loc[countries_filter['GDP per Capita'].idxmax()]

#Print the name of the country
print(wdi[wdi['Country Code']==max_country.name]['Country Name'].iloc[0])
```

```python
#10
#Filter the wdi df to include only rows where the indicator name is "GDP (current US$)"
dataset = wdi.loc[wdi['Indicator Name'] == "GDP (current US$)"]

#Calculate the growth rate in GDP between 1980 and 2010 for all countries
growth_rate = (dataset['2010'] - dataset['1980']) / dataset['1980']

#Filter out rows cause if the the value was 0 or Na, cause it would lead into infinite growth rates
increase = growth_rate[(~dataset['1980'].isna()) & (~dataset['2010'].isna()) & (dataset['1980'] != 0)]

#Find the index of the country with the largest percentage increase
max_increase_index = increase.idxmax()

#Print the name of the country
print(dataset['Country Name'].loc[max_increase_index])
```
